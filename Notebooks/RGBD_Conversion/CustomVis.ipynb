```python
import open3d as o3d
import numpy as np
import matplotlib.pyplot as plt

#Data Path
path = "../../../Data/Tiles/"

from os import listdir
from os.path import isfile, join

allfiles = [f for f in listdir(path) if isfile(join(path, f))]
allObjNames = [f for f in allfiles if (f.find(".obj") != -1)]

index = 1
mesh = o3d.io.read_triangle_mesh(path + allObjNames[index],True)

verts = np.asarray(mesh.vertices)
center = mesh.get_center()
ind = np.where(verts[:,1]==np.max(verts[:,1]))#Assuming it is on an angle and that angle is not 90deg
point = verts[ind,0:2][0,0] - center[0:2]
toa = point[1]/point[0]
theta0 = np.arctan(toa)
theta = theta0-np.pi/4+0.027
mesh.rotate(mesh.get_rotation_matrix_from_xyz((0, 0, -0.3233476477503846)),center=center)
```

```python
def move_forward(vis):
        # This function is called within the o3d.visualization.Visualizer::run() loop
        # The run loop calls the function, then re-render
        # So the sequence in this function is to:
        # 1. Capture frame
        # 2. index++, check ending criteria
        # 3. Set camera
        # 4. (Re-render)
    dep = vis.capture_depth_float_buffer(True)
    plt.imsave("Depth.png",dep)
    vis.capture_screen_image("Image.png",True)
    vis.register_animation_callback(None)
    vis.destroy_window()
    return False
```

```python
#cam_param = o3d.camera.PinholeCameraParameters()
#cam_param.extrinsic=np.identity(4)

vis = o3d.visualization.Visualizer()
vis.create_window(width=1080,height=1080,visible=False)
vis.add_geometry(mesh)
ctr = vis.get_view_control()
ctr.change_field_of_view(step=-13)
#ctr.convert_from_pinhole_camera_parameters(cam_param)
ctr.set_zoom(0.5)

vis.register_animation_callback(move_forward)
vis.run()

del ctr
del vis
```
